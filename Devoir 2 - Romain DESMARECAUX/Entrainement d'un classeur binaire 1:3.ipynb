## Récupérer les données

Pour la récupération des données, nous allons utiliser la fonction **read_csv()** de la librairie *Pandas*. Dans notre cas, nous avons exporté les données d'entraînement et de test de Customers du devoir 1. En plus, nous utiliserons les datasets countryPopulation (le nombre d'habitant par pays) et countryPIB (le PIB par pays).

```python
# importation du module pandas
import pandas as pd 
import numpy as np

#Récupération des données avec la fonction read_csv()
customers = pd.read_csv("Data_traites.csv") 
customers_test = pd.read_csv("Test_traites.csv")
country_population = pd.read_csv("CountryPopulation.csv")
country_PIB = pd.read_csv("CountryGDP.csv")
```

## Création de la fonction *to_binaire()*

Le but de la fonction est de modifier nos deux datasets (customers, customers_test) avec des variables binaires. Les deux variables a modifier sont *revenue* et *gender*.

Avant d'utiliser la fonction to_binaire(), il est important de travailler sur des copies de nos datasets.

```python
#Implémentation de la fonction to_binaire()
def to_binaire(data):
    #Récupérer la valeur moyenne de la variable revenue
    mean = data['revenue'].mean()
    newcolonne = []
    #Tester pour toutes les valeurs de revenue si elles sont < ou > que 'mean'
    for valeur in data["revenue"]:
        #Si la valeur est < que la moyenne alors on remplace la valeur par 0
        if(valeur < mean):
            newcolonne.append(0)
        #Sinon on remplace la valeur par 1
        else:
            newcolonne.append(1)
    data["revenue"] = newcolonne
    newcolonne = []
    #Tester pour toutes les valeurs de la variable 'gender' et la remplacer par 0 pour 'Masc' sinon par 1
    for valeur in data["gender"]:
        if(valeur == 'Masc'):
            valeur = '0'
        else:
            valeur = '1'
        newcolonne.append(pd.to_numeric(valeur, errors = 'coerce'))
    data["gender"] = newcolonne

#Création de copies
customers_copy = customers.copy()
test_copy = customers_test.copy()

#Utilisation de la fonction to_binaire()
to_binaire(customers_copy)
to_binaire(test_copy)

#Vérification du fonctionnement de notre fonction
test_copy.head()
        
```

## Création des échantillonages aléatoire stratifiés


Dans un premier temps, nous créons deux datasets pour nos deux scénarios (avec PIB et sans PIB). Ensuite, nous créerons nos échantillons avec de différentes tailles (2000, 4000, 8000).

Pour rappel, l'échantillonages aléatoire stratifiés est une méthode d'échantillonnage qui garantit que les fréquences relatives des niveaux (classes) d'une variable ayant une stratification spécifique sont conservées dans l’échantillon.

### Merge les datasets CountryPopulation et CountryGPD

Pour les deux scénarios, nous mergerons le dataset *CountryPopulation*.

```python
from sklearn.preprocessing import FunctionTransformer

#Création de la fonction pour merge
def add_extra_features(data, add_PIB = True):
    #Merge CountryPopulation à data par rapport à la variable 'country'
    data = data.merge(country_population, on = 'country')
    if(add_PIB):
        #Merge CountryGDP à data par rapport à la variable 'country'
        data = data.merge(country_PIB, on = 'country')
        return data
    else:
        #Ne pas merge CountryGDP
        return data

#Utilisation de la fonction pour merge
#Scénario 1 (dataset d'entraînement)
customers_extra_noPIB = add_extra_features(customers_copy, False)
#Scénario 2 (dataset d'entraînement)
customers_extra_PIB = add_extra_features(customers_copy)

#Scénario 1 (dataset de test)
testCustomers_extra_noPIB = add_extra_features(test_copy, False)
#Scénario 2 (dataset de test)
testCustomers_extra_PIB = add_extra_features(test_copy)

#Vérification de nos merge
customers_extra_noPIB.head()
customers_extra_PIB.head()
testCustomers_extra_noPIB.head()
testCustomers_extra_PIB.head()
```

## Entraîner et tester par validation croisée (k = 3 , k = 10)

La validation croisée est une méthode d’estimation de fiabilité d’un modèle fondée sur une technique d’échantillonnage.

Tout d'abord, il faut séparer notre dataset en deux (données et résultats). Une partie prendra toute les variables à part 'revenue' et 'country'. Nous n'utilisons pas 'country' car les variables PIB et Population donnent assez d'informatiion du pays du client.

Ensuite, nous séparons les données en données de d'entraînement (80%) et de test (20%).

Enfin, nous utiliserons la cross validation avec un DecisionTreClassfier afin d'entraîner et d'estimer la fiabilité de nos échantillonages.

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score

#Implémentation de la fonction pour vérifier la fiabilité du modéle
def score_val_croisee(data, k):
    #Séparation des données
    X_PIB = data.drop(["revenue", "country"],axis=1) 
    y_PIB = data.revenue

    #Création des données d'entraînement et de test
    X_train, X_test, y_train, y_test = train_test_split(X_PIB, y_PIB, test_size=0.2, random_state=1)

    #Création du classifier
    clf = DecisionTreeClassifier(random_state=42)
    sk_folds = StratifiedKFold(n_splits = k)

    #Application de la validation croisée
    scores = cross_val_score(clf, X_train, y_train, cv = sk_folds)

    #Retourne les scores
    print("Cross Validation Scores: ", scores)
    print("Average CV Score: ", scores.mean())
    print("Number of CV Scores used in Average: ", len(scores))
    
    return scores.mean()

#k = 3
k3 = []
#scénario 1
k3.append(score_val_croisee(echantillon_customers_noPIB2000, 3))
k3.append(score_val_croisee(echantillon_customers_noPIB4000, 3))
k3.append(score_val_croisee(echantillon_customers_noPIB8000, 3))
#scénario 2
k3.append(score_val_croisee(echantillon_customers_PIB2000, 3))
k3.append(score_val_croisee(echantillon_customers_PIB4000, 3))
k3.append(score_val_croisee(echantillon_customers_PIB8000, 3))

#k = 10
k10 = []
#scénario 1
k10.append(score_val_croisee(echantillon_customers_noPIB2000, 10))
k10.append(score_val_croisee(echantillon_customers_noPIB4000, 10))
k10.append(score_val_croisee(echantillon_customers_noPIB8000, 10))

#scénario 2
k10.append(score_val_croisee(echantillon_customers_PIB2000, 10))
k10.append(score_val_croisee(echantillon_customers_PIB4000, 10))
k10.append(score_val_croisee(echantillon_customers_PIB8000, 10))


#Résultats moyen de k = 3 et k = 10
print(sum(k3) / len(k3))
print(sum(k10) / len(k10))
```

## Mesurer les métriques de précision, rappel et F1-Mesure.

Pour mesurer les métriques de précision, rappel et F1-Mesure, nous utiliserons la fonction **classification_report()** de *sklearn*. Avant tout, nous entraînerons notre classifier sur les données d'entraînement pour ensuite prédire les valeurs de test. 

Nous trouverons le meilleur modèle sur un seul échantillon (la représentation graphique représentera tout les échantillons)

```python
#Séparation des données
X_PIB = echantillon_customers_noPIB2000.drop(["revenue", "country"],axis=1) 
y_PIB = echantillon_customers_noPIB2000.revenue

#Création des données d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_PIB, y_PIB, test_size=0.2, random_state=1)


#Création du classifier
clf = DecisionTreeClassifier(random_state=42)

#Entraîner le classifier
clf = clf.fit(X_train, y_train)

#Prédire les valeurs de test
y_pred = clf.predict(X_test)

#Utilisation de la fonction classification_report
print(sklearn.metrics.classification_report(y_test, y_pred))

#Les résultats sont satisfaisant avec des précisions de 75% (0) et 63% (1)
```

## Optimiser les hyper-paramètres (GridSearchCV)

Le but de cette partie est d'optimisiser les hyper-paramètres à l'aide de GridSearchCV afin d'essayer de trouver un meilleur modèle avec des résultats plus performant.

Nous prendrons 2 hyper-paramètres, par soucis de temps de calcule, la profondeur de l’arbre et le nombre minimal d’observations pour une segmentation d’un nœud.

Dans un premier temps, nous prendrons la profondeur de l'arbre de 2 à 19 car partir de 1 serait inutile et nous voulons pas arriver à des temps de calcul trop longdonc nous nous arrêtons à 19. Et pour le nombre minimal d’observations pour une segmentation d’un nœud nous prendrons les valeurs 10, 15 et 20.

Pour le nombre de blocs (k) nous prendrons la meilleure valeur de k trouvée en 2.1.3, d'où k = 10.

```python
from sklearn.model_selection import GridSearchCV

#Initialisation des hyper-paramètres
params = {'max_depth': list(range(2,20)), 'min_samples_split': [10,15,20]}

#Utilisation de GridSearchCV
grid_search_cv = GridSearchCV(DecisionTreeClassifier(random_state=42),
                             params, n_jobs=-1, verbose=1, cv=10)

#Entraînement du GridSearchCV
grid_search_cv.fit(X_train, y_train)

#Retourne les meilleurs valeurs pour les hyper-paramètres 
print(grid_search_cv.best_estimator_)
```

## Tester le meilleur modèle

Après avoir trouver le meilleur modèle, nous allons le tester sur nos données de test puis mesurer les métriques de précision.

```python
#Préparation des données de test (vu auparavant pour les données d'entraînement)
X_Test = testCustomers_extra_noPIB.drop(["revenue", "country"],axis=1)
y_Test = testCustomers_extra_noPIB.revenue

#Prédit les données de test grâce au GridSearchCV entraîné
grid_prediction = grid_search_cv.predict(X_Test) 

#Mesure les métriques de précision
print(sklearn.metrics.classification_report(y_Test, grid_prediction))
```

On observe une net amélioration de nos résultats passant de 75% à 84% pour les 0 et de 63% à 66% pour les 1. On peut donc en conclure que nous bien executé notre GridSearchCV afin d'avoir un meilleur modèle.
